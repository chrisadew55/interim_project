# Q5 - What is the relationship between store trading duration and revenue?

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

# Overall trade duration vs average revenue

```python
df = pd.read_csv(r"C:\Users\inbox\Desktop\INTERIM PROJECT\CSV files\Q5 avg_rev csv.csv")
print(df)

df.corr()
print(df.corr())

sizes =[250]

plt.scatter(df.YearOpened, df.avg_rev, s=sizes, color='black', marker='o')
plt.xlabel('Store opening year') 
plt.ylabel('Average revenue')
plt.title('Trade duration vs average revenue')
plt.show()
```

# Mountain bikes contributed to the highest average revenue in 1999 overall

```python
mountain = pd.read_csv(r"C:\Users\inbox\Desktop\INTERIM PROJECT\CSV files\mountainbikes_avg_rev_onlycsv.csv")
print(mountain)

mountain.corr()
print(mountain.corr())

sizes =[250]
plt.scatter(mountain.YearOpened, mountain.max_rev, s=sizes, color='plum', marker='o', alpha=0.5)
plt.xlabel('Trading year') 
plt.ylabel('Max average revenue')
plt.title('Mountain bikes')
plt.show()
```

# Touring bikes

```python
touring = pd.read_csv(r"C:\Users\inbox\Desktop\INTERIM PROJECT\CSV files\touringbikes_avg_rev_onlycsv.csv")
print(touring)

touring.corr()
print(touring.corr())

sizes =[250]
plt.scatter(touring.YearOpened, touring.max_rev, s=sizes, color='darkcyan', marker='o', alpha=0.5)
plt.xlabel('Trading year') 
plt.ylabel('Max average revenue')
plt.title('Touring bikes')
plt.show()
```

# Road bikes

```python
road = pd.read_csv(r"C:\Users\inbox\Desktop\INTERIM PROJECT\CSV files\roadbikes_avg_rev_onlycsv.csv")
print(road)

road.corr()
print(road.corr())

sizes =[250]
plt.scatter(road.YearOpened, road.max_rev, s=sizes, color='gold', marker='o', alpha=0.5)
plt.xlabel('Trading year') 
plt.ylabel('Max average revenue')
plt.title('Road bikes')
plt.show()
```

# Bikes overlapped

```python
sizem = [150]
sizet = [150]
sizer = [150]

plt.scatter(mountain.YearOpened, mountain.max_rev, s=sizem, color='plum', facecolors='none', marker='<', label='M')
plt.scatter(touring.YearOpened, touring.max_rev, s=sizet, color='darkcyan', facecolors='none', marker='>', alpha=0.3, label='T')
plt.scatter(road.YearOpened, road.max_rev, s=sizer, color='gold', facecolors='none', marker='^', alpha=0.5, label='R')
plt.xlabel('Trading year') 
plt.ylabel('Max average revenue')
plt.title('All specialty bike types')
plt.legend()
plt.show()
```
